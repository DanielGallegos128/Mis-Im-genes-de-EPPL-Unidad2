#### 📝 Redacción del problema
Desarrollar un programa interactivo en Python que permita calcular tres magnitudes geométricas básicas a partir del radio ingresado por el usuario: el área de un círculo, el perímetro (circunferencia) de un círculo y el volumen de una esfera. El programa debe mostrar un menú numérico que se repita hasta que el usuario decida salir, solicitando el radio únicamente cuando se selecciona una opción de cálculo. No se requiere validación de datos ni uso de programación orientada a objetos.

### Formulas

1. 🔵 Área del círculo $A=π⋅r2$
2. 🟢 Perímetro del círculo $P=2⋅π⋅r$
3. 🟣 Volumen de la esfera $V=4/3⋅π⋅r3$

```python
# Importamos la constante matemática pi
import math

def areacirculo():
    print("\n🔵 Cálculo del Área del Círculo")
    radio = float(input("📏 Ingresa el radio del círculo: "))
    area = math.pi * (radio ** 2)
    print(f"✅ El área del círculo con radio {radio} es: {area:.2f} unidades²\n")

def perimetrocirculo():
    print("\n🟢 Cálculo del Perímetro del Círculo")
    radio = float(input("📏 Ingresa el radio del círculo: "))
    perimetro = 2 * math.pi * radio
    print(f"✅ El perímetro del círculo con radio {radio} es: {perimetro:.2f} unidades\n")

def volumenesfera():
    print("\n🟣 Cálculo del Volumen de la Esfera")
    radio = float(input("📏 Ingresa el radio de la esfera: "))
    volumen = (4/3) * math.pi * (radio ** 3)
    print(f"✅ El volumen de la esfera con radio {radio} es: {volumen:.2f} unidades³\n")

def menu():
    print("✨ Bienvenido al Programa de Cálculos Geométricos ✨")

    # Bucle infinito que se mantiene activo hasta que el usuario elija salir
    while True:
        print("\n📋 Menú Principal:")
        print("1️⃣ Calcular área del círculo")
        print("2️⃣ Calcular perímetro del círculo")
        print("3️⃣ Calcular volumen de la esfera")
        print("4️⃣ Salir")

        opcion = int(input("\n👉 Selecciona una opción (1-4): "))

        if opcion == 1:
            areacirculo()
        elif opcion == 2:
            perimetrocirculo()
        elif opcion == 3:
            volumenesfera()
        elif opcion == 4:
            print("\n👋 ¡Gracias por usar el programa! Hasta pronto. 🚀")
            print("Programa realizado por: Daniel Gallegos NL = 20")
            break  # Sale del bucle while

        else:
            print("\n⚠️ Opción no válida. Por favor, elige un número del 1 al 4.")

# Ejecutamos el programa
if __name__ == "__main__":
    menu()
```
